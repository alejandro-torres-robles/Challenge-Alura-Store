```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print(f"Ingreso total tienda 1: {ingreso_tienda1}")
print(f"Ingreso total tienda 2: {ingreso_tienda2}")
print(f"Ingreso total tienda 3: {ingreso_tienda3}")
print(f"Ingreso total tienda 4: {ingreso_tienda4}")
```

```python
# Sumar ingresos por tienda
ingresos = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

# Etiquetas
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Gráfico circular
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title("Distribución del ingreso total por tienda")
plt.axis("equal")
plt.show()
```

```python
# Conteo por categoría para cada tienda
categorias_tienda1 = tienda["Categoría del Producto"].value_counts()
categorias_tienda2 = tienda2["Categoría del Producto"].value_counts()
categorias_tienda3 = tienda3["Categoría del Producto"].value_counts()
categorias_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Mostrar las categorías más vendidas por tienda
print("Categorías más populares - Tienda 1:")
print(categorias_tienda1.head(), end="\n\n")

print("Categorías más populares - Tienda 2:")
print(categorias_tienda2.head(), end="\n\n")

print("Categorías más populares - Tienda 3:")
print(categorias_tienda3.head(), end="\n\n")

print("Categorías más populares - Tienda 4:")
print(categorias_tienda4.head())
```

```python
# Calcular la calificación promedio por tienda
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

# Mostrar los resultados
print(f"Valoración media Tienda 1: {promedio_tienda1:.2f}")
print(f"Valoración media Tienda 2: {promedio_tienda2:.2f}")
print(f"Valoración media Tienda 3: {promedio_tienda3:.2f}")
print(f"Valoración media Tienda 4: {promedio_tienda4:.2f}")
```

```python
def resumen_productos(df, nombre):
    conteo = df["Producto"].value_counts()
    print(f"{nombre} - Más vendido: {conteo.idxmax()} ({conteo.max()} ventas)")
    print(f"{nombre} - Menos vendido: {conteo.idxmin()} ({conteo.min()} venta)")
    print()

resumen_productos(tienda, "Tienda 1")
resumen_productos(tienda2, "Tienda 2")
resumen_productos(tienda3, "Tienda 3")
resumen_productos(tienda4, "Tienda 4")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Función para graficar productos más y menos vendidos
def graficar_productos_mas_menos_vendidos(df, nombre_tienda):
    # Contar productos vendidos
    conteo_productos = df["Producto"].value_counts()

    # Crear figura con 2 subgráficos
    fig, axs = plt.subplots(1, 2, figsize=(14, 5))

    # Top 5 productos más vendidos
    conteo_productos.head(5).plot(kind='bar', ax=axs[0], color='skyblue')
    axs[0].set_title(f'Top 5 más vendidos - {nombre_tienda}')
    axs[0].set_ylabel("Cantidad de ventas")
    axs[0].tick_params(axis='x', rotation=45)

    # Bottom 5 productos menos vendidos
    conteo_productos.tail(5).plot(kind='bar', ax=axs[1], color='salmon')
    axs[1].set_title(f'Bottom 5 menos vendidos - {nombre_tienda}')
    axs[1].set_ylabel("Cantidad de ventas")
    axs[1].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

# Ejecutar para cada tienda
graficar_productos_mas_menos_vendidos(tienda, "Tienda 1")
graficar_productos_mas_menos_vendidos(tienda2, "Tienda 2")
graficar_productos_mas_menos_vendidos(tienda3, "Tienda 3")
graficar_productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

```python
# Calcular el envío promedio para cada tienda
envio_prom_t1 = tienda["Costo de envío"].mean()
envio_prom_t2 = tienda2["Costo de envío"].mean()
envio_prom_t3 = tienda3["Costo de envío"].mean()
envio_prom_t4 = tienda4["Costo de envío"].mean()

# Mostrar resultados
print(f"Envío promedio Tienda 1: ${envio_prom_t1:.2f}")
print(f"Envío promedio Tienda 2: ${envio_prom_t2:.2f}")
print(f"Envío promedio Tienda 3: ${envio_prom_t3:.2f}")
print(f"Envío promedio Tienda 4: ${envio_prom_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
costos_envio = [envio_prom_t1, envio_prom_t2, envio_prom_t3, envio_prom_t4]

# Gráfico de línea
plt.plot(tiendas, costos_envio, marker='o', linestyle='-', color='royalblue')
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo promedio ($)")
plt.xlabel("Tienda")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Tras analizar los ingresos, la rotación de productos, las calificaciones de los clientes, el costo de envío y el posicionamiento de cada tienda, se concluye que la Tienda 4 es la opción más conveniente para que el Sr. Juan la venda. Esta tienda presenta consistentemente el menor desempeño: tiene los ingresos más bajos, una valoración promedio negativa por parte de los clientes, y carece de productos o categorías bien posicionadas en el mercado. Aunque su costo de envío es bajo, esto no compensa su bajo volumen de ventas ni su débil potencial de crecimiento. En contraste, las otras tiendas muestran fortalezas claras en al menos uno o más indicadores clave. Por lo tanto, deshacerse de la Tienda 4 permitiría al Sr. Juan obtener liquidez inmediata sin comprometer significativamente la rentabilidad de su negocio, liberando recursos que podría reinvertir estratégicamente en las unidades más sólidas.
